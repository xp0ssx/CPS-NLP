```python
import requests

OLLAMA_URL = "http://127.0.0.1:11434/api/generate"
MODEL_NAME = "qwen2.5:0.5b"


def send_prompt(prompt: str) -> str:
    """Send one prompt to Ollama and return the generated answer."""
    payload = {
        "model": MODEL_NAME,
        "prompt": prompt,
        "stream": False,
    }
    response = requests.post(OLLAMA_URL, json=payload, timeout=120)
    response.raise_for_status()
    return response.json()["response"].strip()
```

```python
answer = send_prompt(
    "Объясни простыми словами, что такое машинное обучение. "
    "Ответь кратко, без лишних оправданий, 3-4 предложения."
)
print(answer)
```

```python
import csv

prompts = [
    "Кратко объясни, что такое машинное обучение.",
    "Объясни разницу между HTTP и HTTPS.",
    "Что такое Docker?",
    "Назови 3 преимущества Python.",
    "Как работает сортировка пузырьком?",
    "Придумай короткий тост для дня рождения.",
    "Объясни, что такое база данных.",
    "Составь план изучения NLP на 1 неделю.",
    "Переведи фразу 'good morning' на русский.",
    "Назови 5 городов России.",
]

rows = []

for prompt in prompts:
    response = send_prompt(prompt)
    rows.append({"prompt": prompt, "response": response})

with open("inference_report.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(file, fieldnames=["prompt", "response"])
    writer.writeheader()
    writer.writerows(rows)
```
